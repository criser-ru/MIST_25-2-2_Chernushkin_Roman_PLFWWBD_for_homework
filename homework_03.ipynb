```python
def count_word_frequency(text):
    text = text.lower()

    for znak in [".", ",", "!", "?"]:
        text = text.replace(znak, "")

    words = text.split()

    frequency = {}

    for word in words:
        frequency[word] = frequency.get(word, 0) + 1

    return frequency


# Пример использования
text = "Это просто пример. Просто очень простой пример текста для примера."
frequency_dict = count_word_frequency(text)
print(frequency_dict)
```

```python
def calculate_price(price, discount=0, tax_rate=0.2):
    price_after_discount = price * (1 - discount / 100)

    final_price = price_after_discount * (1 + tax_rate)

    return final_price


# Пример использования:
result = calculate_price(1000, discount=10)
print(result)   # Цена после скидки 900 → после налога 1080
```
